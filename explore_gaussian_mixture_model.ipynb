```python
# This script is written to explore Gaussian Mixture Models (GMM)
from sklearn import mixture
import pandas as pd
import seaborn as sns
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

df = pd.read_csv('morpho_data.csv')

features = list(df.drop("target", axis=1))
X= df.drop("target", axis=1)

# First fit model PCA
model = PCA(n_components =2)
model.fit(X)
X_2D = model.transform(X)
df['PCA1'] = X_2D[:, 0]
df['PCA2'] = X_2D[:, 1]
sns.lmplot("PCA1", "PCA2", hue = 'target', data = df, fit_reg = False)
plt.show(sns)

model = mixture.GaussianMixture(n_components=3, covariance_type='full')
model.fit(X)
y_gmm = model.predict(X)
df['cluster'] = y_gmm
sns.lmplot("PCA1", "PCA2", data=df, hue='target', col='cluster', fit_reg=False)
plt.show(sns)
```
